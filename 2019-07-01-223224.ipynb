```python
import pandas as pd 
webdata = pd.read_csv('S.csv')
```

Data Cleaning stage, We are cleaning the data with following condition; Borough must be assigned, All Neighborhood related with specific postcode should be alligned next to each other

```python
webdata
webdata.drop(webdata[webdata['Borough']=="Not assigned"].index,axis=0,inplace=True)
webdata.drop(webdata[webdata['Neighbourhood']=="Not assigned"].index,axis=0,inplace=True)
webdata.groupby('Postcode').agg(lambda x:','.join(set(x)))
webdata.loc[webdata['Neighbourhood']=="Not assigned",'Neighbourhood']=webdata.loc[webdata['Neighbourhood']=="Not assigned",'Borough']
webdata.shape
```
